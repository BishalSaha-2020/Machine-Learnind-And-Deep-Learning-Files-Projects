```python
import cv2
#img=cv2.imread('12134015332p.jpg')
import datetime
cap = cv2.VideoCapture(0)
cap.set(3,1280)
cap.set(4,720)
cap.set(40,130)

classNames=[]
classFile = 'coco.names'
with open(classFile,'rt') as f:
   classNames = f.read().rstrip('\n').split('\n')


configPath = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
weightsPath = 'frozen_inference_graph.pb'

net = cv2.dnn_DetectionModel(weightsPath,configPath)
net.setInputSize(320,320)
net.setInputScale(1.0/ 127.5)
net.setInputMean((127.5, 127.5, 127.5))
net.setInputSwapRB(True)

while True:
    success,img = cap.read()
    classIds, confs, bbox = net.detect(img,confThreshold=0.45)
    print(classIds,bbox)
    dt = str(datetime.datetime.now())
  
        

    
 
    # put the dt variable over the
    # video frame
    font = cv2.FONT_HERSHEY_SCRIPT_COMPLEX
    frame = cv2.putText(img, dt,
                        (10, 100),
                        font, 1,
                        (100, 155, 50),
                        4, cv2.LINE_8)
 
    if len(classIds) != 0:
        for classId, confidence,box in zip(classIds.flatten(),confs.flatten(),bbox):
            with open('readmeys.txt', 'a') as f:
                f.write(classNames[classId-1].upper())
                f.write(dt)
            cv2.rectangle(img,box,color=(0,255,0),thickness=2)
            cv2.putText(img,classNames[classId-1].upper(),(box[0]+10,box[1]+30),
           
            cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
            cv2.putText(img,str(round(confidence*100,2)),(box[0]+200,box[1]+30),
            cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
           
        
    cv2.imshow('Video', img)
    print(dt)
    print(classNames[classId-1])
   
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
